```python
# Merging Order_Report and SKU_Master
merged_data=pd.merge(Order_Report, SKU_Master, on='SKU')
merged_data
```

```python
# Merging merged_data with Invoice
merged_data=pd.merge(merged_data, Invoice, left_on='ExternOrderNo', right_on='Order ID', how='left')
merged_data
```

```python
# Merging merged_data with Pincode_Zones
merged_data=pd.merge(merged_data, Pincode_Zones, on='Warehouse Pincode')
merged_data
```

```python
#Checking null values
merged_data.isna().sum()
```

```python
# Checking Duplicate Values
merged_data.duplicated().sum()
```

```python
# dropping duplicates based on specific column
merged_data=merged_data.drop_duplicates(subset=['ExternOrderNo','AWB Code'])
```

```python
# converting weight (g) column into float for merging Rates with merged_data
merged_data['Weight (g)']=merged_data['Weight (g)'].astype(float)
merged_data.info()
```

```python
merged_data['Zone_x'].unique()
```

```python
merged_data['Zone_y'].unique()
```

```python
Rates['Zone'].unique()
```

```python
# coverting Zone_x and Zone_y columns into Uppercase
merged_data['Zone_x_upper']=merged_data['Zone_x'].str.upper()
merged_data['Zone_y_upper']=merged_data['Zone_y'].str.upper()
```

```python
merged_data=pd.merge(merged_data, Rates, left_on=['Zone_y', 'Weight (g)'], right_on=['Zone','Weight Slabs'], how='left')
```

```python
merged_data.isna().sum()
```

```python
#fill null values with default values
default_values={'Zone': 'UNKNOWN_ZONE', 'Weight Slabs':0, 'Forward Fixed Charge':0, 'Forward Additional Weight Slab Charge':0,
               'RTO Fixed Charge':0, 'RTO Additional Weight Slab Charge':0}
```

```python
merged_data=merged_data.fillna(default_values)
```

```python
merged_data.isna().sum()
```

```python
merged_data=merged_data.drop(columns=['Zone_x_upper', 'Zone_y_upper'])
```

```python
# Calculating weight based on weight slab
def calculated_weight(row):
    Weight_slab=row['Weight (g)']/1000
    return Weight_slab*row['Weight Slabs']
```

```python
# Applying function to calculate applicable weight
merged_data['Applicable Weight']=merged_data.apply(calculated_weight, axis=1)
```

```python
# Calculation X's weight and courier comapny weight
merged_data['Weight Difference']=merged_data['Applicable Weight']-merged_data['Charged Weight']
merged_data['Weight Difference']
```

```python
# to calculate expected charge
def calculate_expected_charge(row):
    Weight_slab_charge=row['Forward Fixed Charge']
    return Weight_slab_charge
```

```python
# Applying funtion to calculate Expected Charge
merged_data['Expected Charge']=merged_data.apply(calculate_expected_charge, axis=1)
```

```python
# Empty list to store rows for summary table
Summary_rows=[]
```

```python
# calculating total orders and amounts
Correctly_Charged=merged_data[merged_data['Expected Charge']==merged_data['Billing Amount (Rs.)']]
Over_Charged=merged_data[merged_data['Expected Charge']>merged_data['Billing Amount (Rs.)']]
Under_Charged=merged_data[merged_data['Expected Charge']<merged_data['Billing Amount (Rs.)']]
```

```python
Summary_rows.append({'Category':'Total Orders-Correctly_Charged', 'Count':len(Correctly_Charged), 
                    'Amount (Rs.)':Correctly_Charged['Billing Amount (Rs.)'].sum()})
```

```python
Summary_rows.append({'Category':'Total Orders-Over_Charged', 'Count':len(Over_Charged), 
                     'Amount (Rs.)':Over_Charged['Billing Amount (Rs.)'].sum()})
```

```python
Summary_rows.append({'Category':'Total Orders-Under_Charged', 'Count':len(Under_Charged), 
                     'Amount (Rs.)':Under_Charged['Billing Amount (Rs.)'].sum()})
```

```python
# Create Dataframe from the list of rows
Summary_table=pd.DataFrame(Summary_rows)
Summary_table
```

```python
Order_level_calculation=merged_data[['ExternOrderNo', 'AWB Code', 'Weight (g)', 'Weight Slabs', 'Charged Weight', 'Zone_x', 'Zone_y',
                          'Expected Charge', 'Billing Amount (Rs.)', 'Weight Difference']]
Order_level_calculation
```

```python
Order_level_calculation.columns=['Order ID', 'AWB Number', 'Total weight as per X (KG)',
                        'Total weight as per Courier Company (KG)', 'Weight slab charged by Courier Company (KG)',
                        'Delivery Zone as per X', 'Delivery Zone charged by Courier Company',
                        'Expected Charge as per X (Rs.)', 'Charges Billed by Courier Company (Rs.)',
                        'Difference Between Expected Charges and Billed Charges (Rs.)']
```
